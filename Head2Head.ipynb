```python
import pandas as pd;
import numpy as np;


file_path = 'database_24_25.csv'
nba_data = pd.read_csv(file_path)
```

```python
print("First 10 row of the dataset:")
print(nba_data.head(10));
```

```python
#Show the datatypes and nulls
nba_data.info()
```

```python
#Check for missing values
print('Prints missing value in each column:')
print(nba_data.isnull().sum());
```

```python
print("Displaying dataset to help detect misleading data:")
print(nba_data.describe())
```

```python
#Only include players with minutes played more then 0.
nba_data = nba_data[nba_data['MP'] > 0]
                    

#Preprocessing: checking for missing values after.
print('Prints missing value in each column:')
print(nba_data.isnull().sum());
```

```python
# MatPlot Installation
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Points (PTS)
plt.figure(figsize=(10,6))
sns.histplot(nba_data['PTS'], bins=30, kde=True)
plt.title('Distribution of Player Points per Game')
plt.xlabel('Points per Game')
plt.ylabel('Number of Games')
plt.grid(True)
plt.show()

# Minutes Played (MP)
plt.figure(figsize=(10,6))
sns.histplot(nba_data['MP'], bins=30, kde=True)
plt.title('Distribution of Minutes Played per Game')
plt.xlabel('Minutes Played per Game')
plt.ylabel('Number of Games')
plt.grid(True)
plt.show()

# Assists (AST)
plt.figure(figsize=(10,6))
sns.histplot(nba_data['AST'], bins=30, kde=True)
plt.title('Distribution of Assists per Game')
plt.xlabel('Assists per Game')
plt.ylabel('Number of Games')
plt.grid(True)
plt.show()

# Rebounds (TRB)
plt.figure(figsize=(10,6))
sns.histplot(nba_data['TRB'], bins=30, kde=True)
plt.title('Distribution of Rebounds per Game')
plt.xlabel('Rebounds per Game')
plt.ylabel('Number of Games')
plt.grid(True)
plt.show()

# Turnovers (TOV)
plt.figure(figsize=(10,6))
sns.histplot(nba_data['TOV'], bins=30, kde=True)
plt.title('Distribution of Turnovers per Game')
plt.xlabel('Turnovers per Game')
plt.ylabel('Number of Games')
plt.grid(True)
plt.show()

# Game Score (GmSc)
plt.figure(figsize=(10,6))
sns.histplot(nba_data['GmSc'], bins=30, kde=True)
plt.title('Distribution of Game Performance Score (GmSc)')
plt.xlabel('Game Score')
plt.ylabel('Number of Games')
plt.grid(True)
plt.show()

```

```python
plt.show()

nba_data.to_csv('nba_cleaned.csv', index=False)


```
